```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv("../data/raw/ipo_2010_2021.csv")
df_nifty = pd.read_csv("../data/raw/NIFTY_2010_2021.csv")

print(f"Original shape: {df.shape}")
print("Loaded successfully!")
```

```python
# Profile column is just URLs — no value for ML
df = df.drop(columns=["Profile"])

print("Dropped Profile column")
print(f"New shape: {df.shape}")
print(f"Columns remaining: {list(df.columns)}")
```

```python
# Date is in format '29-07-21' — convert to proper datetime
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%y")

# Extract useful time features from date
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Quarter"] = df["Date"].dt.quarter

print("Date column fixed!")
print(df["Date"].head())
print(f"\nYear range: {df['Year'].min()} to {df['Year'].max()}")
```

```python
# CMP has commas like "2,268.50" — remove commas and convert to float
df["CMP"] = df["CMP"].astype(str).str.replace(",", "", regex=False)
df["CMP"] = pd.to_numeric(df["CMP"], errors="coerce")

print("CMP column fixed!")
print(f"CMP dtype: {df['CMP'].dtype}")
print(f"Sample values: {df['CMP'].dropna().head().values}")
```

```python
# Rename columns to clean snake_case — easier to work with in code
df = df.rename(columns={
    "IPO Name"              : "ipo_name",
    "Issue Size (in crores)": "issue_size_cr",
    "QIB"                   : "qib_subscription",
    "HNI"                   : "hni_subscription",
    "RII"                   : "rii_subscription",
    "Total"                 : "total_subscription",
    "Issue"                 : "issue_price",
    "Listing Open"          : "listing_open",
    "Listing Close"         : "listing_close",
    "Listing Gains(%)"      : "listing_gains_pct",
    "CMP"                   : "cmp",
    "Current Gains (%)"     : "current_gains_pct"
})

print("Columns renamed!")
print(list(df.columns))
```

```python
print("Missing values before treatment:")
print(df.isnull().sum())

# Fill missing subscription values with median
for col in ["qib_subscription", "hni_subscription",
            "rii_subscription", "total_subscription"]:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)
    print(f"Filled {col} nulls with median: {median_val:.2f}")

# Fill missing CMP with listing_close (best approximation)
df["cmp"] = df["cmp"].fillna(df["listing_close"])

print("\nMissing values after treatment:")
print(df.isnull().sum())
```

```python
# Absolute listing gain in rupees
df["listing_gain_abs"] = df["listing_close"] - df["issue_price"]

# Categorize IPOs by listing performance
def categorize_gain(pct):
    if pct < 0:
        return "Loss"
    elif pct < 10:
        return "Weak"
    elif pct < 30:
        return "Moderate"
    elif pct < 50:
        return "Strong"
    else:
        return "Blockbuster"

df["performance_category"] = df["listing_gains_pct"].apply(categorize_gain)

print("New columns added!")
print(df[["ipo_name", "listing_gains_pct",
          "performance_category"]].head(10))
```

```python
# Fix NIFTY date column
df_nifty["Date"] = pd.to_datetime(df_nifty["Date"])

# Rename columns
df_nifty = df_nifty.rename(columns={
    "Open"              : "nifty_open",
    "High"              : "nifty_high",
    "Low"               : "nifty_low",
    "Close"             : "nifty_close",
    "Shares Traded"     : "shares_traded",
    "Turnover (Rs. Cr)" : "turnover_cr"
})

# Calculate 30-day rolling return on Nifty
df_nifty = df_nifty.sort_values("Date").reset_index(drop=True)
df_nifty["nifty_30d_return"] = df_nifty["nifty_close"].pct_change(30) * 100

print("NIFTY dataset cleaned!")
print(df_nifty.head(3))
```

```python
# Match each IPO date with the Nifty 30-day return
# This tells us: what was the market mood when this IPO listed?
df = df.sort_values("Date").reset_index(drop=True)

df_merged = pd.merge_asof(
    df,
    df_nifty[["Date", "nifty_close", "nifty_30d_return"]],
    on="Date",
    direction="nearest"
)

print(f"Merged shape: {df_merged.shape}")
print(f"\nSample — IPO with market context:")
df_merged[["ipo_name", "listing_gains_pct",
           "nifty_close", "nifty_30d_return"]].head(5)
```

```python
os.makedirs("../data/cleaned", exist_ok=True)

df_merged.to_csv("../data/cleaned/ipo_cleaned.csv", index=False)

print("Cleaned data saved!")
print(f"Final shape: {df_merged.shape}")
print(f"Final columns: {list(df_merged.columns)}")
```

```python
# Load the cleaned file
df = pd.read_csv("../data/cleaned/ipo_cleaned.csv")

print("Before fixing zeros:")
print(f"Rows with issue_price = 0: {(df['issue_price'] == 0).sum()}")
print(f"Rows with issue_size_cr = 0: {(df['issue_size_cr'] == 0).sum()}")

# Remove rows where issue_price or issue_size_cr is 0
# These are incomplete records — useless for ML
df = df[df["issue_price"] > 0]
df = df[df["issue_size_cr"] > 0]

print("\nAfter fixing zeros:")
print(f"Rows with issue_price = 0: {(df['issue_price'] == 0).sum()}")
print(f"Rows with issue_size_cr = 0: {(df['issue_size_cr'] == 0).sum()}")
print(f"\nFinal shape after cleaning: {df.shape}")
```

```python
df.to_csv("../data/cleaned/ipo_cleaned.csv", index=False)
print("Cleaned file updated and saved!")
```
